```python
import pandas as pd
```

```python
df = pd.read_csv(
    r'C:\Users\daviw\OneDrive\Projetos\Python\Segurança Pública\data\raw\obitos_por_arma_de_fogo_tabnet_2023.csv',
    encoding='ISO-8859-1',
    sep=';',  # Trocar por ';' se necessário
    skiprows=4,
    skipfooter=11,
    engine='python'  # Necessário para skipfooter funcionar
)

df
```

```python
# Dados da tabela (códigos das UFs do TabNet) com a abreviatura
dados = {
    'Código': [11, 12, 13, 14, 15, 16, 17, 21, 22, 23, 24, 25, 26, 27, 28, 29, 31, 32, 33, 35, 41, 42, 43, 50, 51, 52, 53],
    'UF': ['Rondônia', 'Acre', 'Amazonas', 'Roraima', 'Pará', 'Amapá', 'Tocantins', 'Maranhão', 'Piauí', 'Ceará', 'Rio Grande do Norte', 'Paraíba', 'Pernambuco', 'Alagoas', 'Sergipe', 'Bahia', 'Minas Gerais', 'Espírito Santo', 'Rio de Janeiro', 'São Paulo', 'Paraná', 'Santa Catarina', 'Rio Grande do Sul', 'Mato Grosso do Sul', 'Mato Grosso', 'Goiás', 'Distrito Federal'],
    'Sigla': ['RO', 'AC', 'AM', 'RR', 'PA', 'AP', 'TO', 'MA', 'PI', 'CE', 'RN', 'PB', 'PE', 'AL', 'SE', 'BA', 'MG', 'ES', 'RJ', 'SP', 'PR', 'SC', 'RS', 'MS', 'MT', 'GO', 'DF']
}

# Criando o DataFrame
df_ufs = pd.DataFrame(dados)
df_ufs['Código'] = df_ufs['Código'].astype(str)
```

```python
df['cod mun'] = df['Município'].str.split(' ').str[0]
df['municipio'] = df['Município'].str.split(' ').str[1:].str.join(' ')
df['cod_uf'] = df['cod mun'].astype(str).str[:2]
df['uf'] = df['cod_uf'].map(df_ufs.set_index('Código')['Sigla'])
df['chave_uf'] = df['uf'] + df['municipio'].str.lower()
df = df.rename(columns={'Óbitos_p/Ocorrênc': 'obitos'})

df
```

```python
df.to_csv(r'C:\Users\daviw\OneDrive\Projetos\Python\Segurança Pública\data\processed\obitos_datasus.csv')
```

```python
df_obitos_arma_fogo_2023_uf = df.groupby(['uf'])['obitos'].sum().reset_index()
df_obitos_arma_fogo_2023_uf
```
